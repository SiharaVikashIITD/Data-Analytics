#The .value_counts()  
function in pandas is used to count the occurrences of each unique value in a Series or column of a DataFrame. It returns a Series containing counts of unique values, sorted in descending order by default.


Counting Values: It counts how many times each unique value appears in the column.

Sorting: The counts are sorted in descending order, with the most frequent values at the top.

Data Type: The result is a Series, where the index contains the unique values, and the corresponding values are the counts.

#plt.xticks(rotation=45)

This line rotates the labels on the x-axis by 45 degrees.
It is particularly useful when the labels are long or when there are many categories, as it helps prevent overlapping and makes the labels easier to read.

#plt.tight_layout()

This function adjusts the layout of the plot to ensure that all elements, such as titles, labels, and tick marks, fit within the figure neatly without overlapping.
It automatically adjusts the subplot parameters to give some padding and avoid clipping of plot elements, making the plot look more polished and professional.

#5.	What proportion of clients have credit in default?

```python
# Counting the number of clients with credit in default
clients_with_default = dataset[dataset['default'] == 'yes'].shape[0]
# Total number of clients in the dataset
total_clients = dataset.shape[0]
# Calculating the proportion of clients with credit in default
proportion_with_default = (clients_with_default / total_clients) * 100
# Printing the result
print(f"Proportion of clients with credit in default: {proportion_with_default:.2f}%")
```

#What is the distribution of average yearly balance among the clients?

We creates hist for numerical valuse and bar for catagorical data

```python
plt.figure(figsize=(10,6))
dataset['balance'].plot(kind='hist',edgecolor='black')
plt.title('distribution of average yearly balance among the clients')
plt.xlabel('average yearly balance')
plt.ylabel('No. of Clients')
plt.tight_layout()
plt.show()
print(dataset['balance'].value_counts())
```

```python
balances = dataset['balance']

# Binning the balance data into categories
bins = [-float('inf'), 0, 500, 1000, 5000, 10000, float('inf')]
labels = ['< 0', '0 - 500', '500 - 1000', '1000 - 5000', '5000 - 10000', '> 10000']
balance_categories = pd.cut(balances, bins=bins, labels=labels)

# Counting the number of clients in each category
balance_counts = balance_categories.value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
balance_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Average Yearly Balance Among Clients')
plt.ylabel('')  # Hides the y-label, which is not necessary for pie charts
plt.show()
print(balance_counts)
```

#17.	Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?

```python
# Step 1: Convert 'y' to binary
dataset['y_binary'] = dataset['y'].apply(lambda x: 1 if x == 'yes' else 0)

# Step 2: Encode categorical variables, including 'y_binary'
df_encoded = pd.get_dummies(dataset, drop_first=True)

# Step 3: Compute the correlation matrix
corr_matrix = df_encoded.corr()

# Step 4: Plot the correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
#plt.subplots_adjust(left=0.2, right=0.95, top=0.95, bottom=0.2) # adjustments for proper labeling and figure
plt.show()

# Step 5: Analyze specific correlations with the target variable 'y_binary'
correlations = corr_matrix['y_binary'].sort_values(ascending=False)
print(correlations)
```
